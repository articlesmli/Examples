JAX: Best for high-performance research and understanding the raw math.

```python
import jax
import jax.numpy as jnp
from jax import grad, jit, vmap
import numpy as np

# 1. Setup Data in JAX format
celsius = jnp.array([-40, -10, 0, 8, 15, 22, 38], dtype=jnp.float32)
fahrenheit = jnp.array([-40, 14, 32, 46, 59, 72, 100], dtype=jnp.float32)

# 2. Initialize Parameters (Weights and Bias)
# In JAX, we manage variables manually
key = jax.random.PRNGKey(0)
w = jax.random.normal(key, (1,))
b = jax.random.normal(key, (1,))

# 3. Define the Model and Loss
def predict(w, b, inputs):
    return w * inputs + b

def loss_fn(params, inputs, targets):
    w, b = params
    preds = predict(w, b, inputs)
    return jnp.mean((preds - targets) ** 2)

# 4. Training Loop (Using JAX's gradient engine)
@jit # This makes the code run super fast (Just-In-Time compilation)
def update(w, b, x, y, lr=0.0001):
    grads = grad(loss_fn)((w, b), x, y)
    w = w - lr * grads[0]
    b = b - lr * grads[1]
    return w, b

print("Training JAX model...")
for i in range(1000):
    w, b = update(w, b, celsius, fahrenheit)
    if i % 200 == 0:
        current_loss = loss_fn((w, b), celsius, fahrenheit)
        print(f"Iteration {i}: Loss {current_loss}")

# 5. Result
test_celsius = 100.0
prediction = predict(w, b, test_celsius)
print(f"\nJAX Prediction for 100°C: {prediction[0]:.2f}°F")
```
