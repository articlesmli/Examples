```python
"""
Bash
"""
# Install Minikube
curl -LO https://storage.googleapis.com/minikube/releases/latest/minikube-linux-amd64

# no reaction
sudo install minikube-linux-amd64 /usr/local/bin/minikube

# Verify
minikube version

# Set up the cluster 
minikube start --driver=docker

# Check the status
kubectl get nodes
```

```python
"""
Bash
""" 
nano deployment.yaml
```

```python
"YAML"

# Put it under theprevious code 

apiVersion: apps/v1
kind: Deployment
metadata:
  name: molecular-predictor
spec:
  replicas: 2
  selector:
    matchLabels:
      app: predictor
  template:
    metadata:
      labels:
        app: predictor
    spec:
      containers:
      - name: jax-model
        image: nginx # Using a placeholder image to test
        ports:
        - containerPort: 80

"""
Save and Exit: * Press Ctrl + O then Enter (to write the file).

Press Ctrl + X (to exit).
"""
```

```python
"""
Bash
"""
kubectl apply -f deployment.yaml

# Check the status
kubectl get pods

# More information for the status of the installlibraries
kubectl describe pod molecular-predictor

# Open new nano
nano service.yaml
```

```python
"""
YAML
"""
apiVersion: v1
kind: Service
metadata:
  name: predictor-service
spec:
  selector:
    app: predictor # This MUST match the label in your deployment.yaml
  ports:
    - protocol: TCP
      port: 80         # The port other apps will use to call the service
      targetPort: 80   # The port your container is listening on
  type: NodePort       # This makes the service accessible on your MyAlien machine
```

```python
"""
Bash
"""
kubectl apply -f service.yaml 
# Must see service/predictor-service created message

# Open the browser
minikube service predictor-service

"""
Welcome to nginx!
If you see this page, nginx is successfully installed and working.
"""
```
