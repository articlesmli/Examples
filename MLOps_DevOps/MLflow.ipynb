```python
"""Paste it in nano"""

import mlflow

# Add this line!
mlflow.set_tracking_uri("sqlite:///mlflow.db")
mlflow.set_experiment("Bio_Model_Test")

with mlflow.start_run():
    mlflow.log_param("status", "working")
    mlflow.log_metric("test_score", 100)
    print("Data saved to sqlite database!")


""" Ctrl+O, Enter, Ctrl+X"""
```

```python
"""Bash"""
# Run the code
python3 train.py

# Start the UI, telling it specifically to look at that same database file: 
mlflow ui --backend-store-uri sqlite:///mlflow.db
```

# Why this is a "GCP" Lesson
In a professional Google Cloud setup, you wouldn't use a local file like mlflow.db. Instead:

You would create a Cloud SQL instance (GCP's database service).

Your tracking URI would look like postgresql://user:password@ip_address/dbname.

The UI would connect to that same remote database.

By using sqlite:///mlflow.db on your PC now, you are practicing exactly how "Remote Tracking" works—just on a smaller scale.
